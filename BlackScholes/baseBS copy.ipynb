```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
from IPython.display import display, clear_output
import torch.optim as optim
from mpl_toolkits.mplot3d import Axes3D  
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)
```

```python
# wiki page: https://en.wikipedia.org/wiki/Black%E2%80%93Scholes_equation
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        nNeurons = 20
        self.fc1 = nn.Linear(2, nNeurons)  # 5*5 from image dimension
        self.fc2 = nn.Linear(nNeurons, nNeurons)
        self.fc3 = nn.Linear(nNeurons, nNeurons)
        self.fc4 = nn.Linear(nNeurons, nNeurons)
        self.fc5 = nn.Linear(nNeurons, nNeurons)
        self.fc6 = nn.Linear(nNeurons, nNeurons)
        self.fc7 = nn.Linear(nNeurons, 1)
        # self.sigma = 1
        self.sigma = 0.1
        # self.r = 1
        self.r = 0.04
        self.K = 0.25
        self.nt = 75
        self.nS = 75
        t = np.linspace(0, 1, self.nt, endpoint=True)
        S = np.linspace(0, 1, self.nS, endpoint=True)
        self.tt, self.SS = np.meshgrid(t, S)
        tInput = self.tt.reshape(-1, 1)
        SInput = self.SS.reshape(-1, 1)
        self.plotX = torch.tensor(
            np.concatenate((tInput, SInput), axis=1), dtype=torch.float32
        ).to(device)

        self.nrand = 100

        self.epochs = 0
        self.losshist = []

    def train(self, x, iterations, lr=1e-5):
        self.optimizer = torch.optim.Adam(self.parameters(), lr=lr)
        itend = self.epochs + iterations
        nplots = 20
        plotEpoch = iterations / nplots
        while self.epochs < itend:
            self.epochs += 1
            loss = self.losses(x)
            loss.backward()
            self.optimizer.step()
            self.optimizer.zero_grad()
            # self.losshist.append(loss.item())
            self.losshist.append(self.losslist)
            if self.epochs % plotEpoch == 0:
                clear_output(wait=True)
                print(f"Epoch: {self.epochs}, loss: {self.losslist}")
                self.plot()
                self.plotLossHist()

    def train2(self, x, iterations):
        self.optimizer = optim.LBFGS(
            self.parameters(),
            max_iter=100,
            tolerance_grad=1e-9,
            line_search_fn="strong_wolfe",
            history_size = 1000
        )
        itend = self.epochs + iterations
        nplots = 20
        plotEpoch = iterations / nplots
        while self.epochs < itend:
            self.epochs += 1

            def closure():
                self.optimizer.zero_grad()
                self.loss = self.losses(x)
                # self.losslist = self.losses(x)
                # self.loss = torch.sum(self.losslist)
                self.loss.backward()
                return self.loss

            self.optimizer.step(closure)
            # self.losshist.append(self.losslist.item())
            self.losshist.append(self.losslist)
            if self.epochs % plotEpoch == 0:
                clear_output(wait=True)
                print(f"Epoch: {self.epochs}, loss: {self.losslist}")
                self.plot()
                self.plotLossHist()

    def losses(self, X):
        lossPDE = self.lossPDE(X)
        lossBC1 = self.lossBC1()
        lossBC2 = self.lossBC2()
        lossBC3 = self.lossBC3()

        loss = lossPDE+lossBC1+lossBC2+lossBC3
        self.losslist = [loss.item(),lossPDE.item(),lossBC1.item(),lossBC2.item(),lossBC3.item()]

        return loss

    def lossPDE(self, X):
        y = self.forward(X)
        # y = self.applyBoundary(X,y)
        t = X[:, 0]
        S = X[:, 1]
        V = y.view(-1)
        V_X = torch.autograd.grad(
            V, X, grad_outputs=torch.ones_like(V), retain_graph=True, create_graph=True
        )[0]
        V_t = V_X[:, 0]
        V_S = V_X[:, 1]
        V_SS = torch.autograd.grad(
            V_S, X, grad_outputs=torch.ones_like(V_S), create_graph=True
        )[0][:, 1]

        pde = (
            V_t
            + 1 / 2 * (self.sigma**2) * (S**2)
            + V_SS
            + self.r * S * V_S
            - self.r * V
        )
        loss = torch.mean(torch.square(pde))

        return loss

    def lossBC1(self):
        t = torch.rand(self.nrand, device=device).view(-1, 1)
        S = torch.zeros_like(t, device=device).view(-1, 1)
        X = torch.concat((t, S), dim=1)
        XTensor = X.clone().requires_grad_(True)
        y = self.forward(XTensor)
        y = self.applyBoundary(X,y)
        loss = torch.mean(torch.square(y))
        return loss

    def lossBC2(self):
        t = torch.rand(self.nrand, device=device).view(-1, 1)
        S = torch.ones_like(t, device=device).view(-1, 1)
        X = torch.concat((t, S), dim=1)
        XTensor = X.clone().requires_grad_(True)
        y = self.forward(XTensor)
        y = self.applyBoundary(X,y)
        loss = torch.mean(torch.square(y - (S-self.K)))
        return loss

    def lossBC3(self):
        S = torch.rand(self.nrand, device=device).view(-1, 1)
        t = torch.zeros_like(S, device=device).view(-1, 1)
        X = torch.concat((t, S), dim=1)
        XTensor = X.clone().requires_grad_(True)
        y = self.forward(XTensor).view(-1, 1)
        y = self.applyBoundary(X,y)
        target = torch.max(S-self.K,torch.zeros_like(S))
        error = y-target
        loss = torch.mean(torch.square(error))
        return loss

    def forward(self, x):
        x = F.tanh(self.fc1(x))
        x = F.tanh(self.fc2(x))
        x = F.tanh(self.fc3(x))
        x = F.tanh(self.fc4(x))
        x = F.tanh(self.fc5(x))
        x = F.tanh(self.fc6(x))
        x = self.fc7(x)
        return x

    def applyBoundary(self, X, y):
        t = X[:,0].view(-1,1)
        S = X[:,1].view(-1,1)
        y = y*S*(1-S)*t*8+F.relu(S-self.K)
        return y.view(-1, 1)

    def plot(self):
        y = self.forward(self.plotX)
        plotY = self.applyBoundary(self.plotX, y)
        yplot = plotY.cpu().detach().numpy()
        ncontours = 30
        plt.figure()
        plt.contourf(
            self.tt,
            self.SS,
            yplot.reshape(self.nS, self.nt),
            cmap="jet",
            levels=ncontours,
        )
        plt.xlabel("t")
        plt.ylabel("S")
        plt.title(f"Epoch = {self.epochs}")
        plt.colorbar()
        plt.show()

        fig = plt.figure()
        ax = fig.add_subplot(111, projection="3d")

        surf = ax.plot_surface(
            self.tt, self.SS, yplot.reshape(self.nS, self.nt), cmap="jet"
        )

        fig.colorbar(surf)
        ax.view_init(20, 200, 0)
        ax.set_xlabel("t")
        ax.set_ylabel("S")
        ax.set_zlabel("V")
        plt.title(f"Epoch = {self.epochs}")
        plt.show()

    def plotLossHist(self):
        plt.figure()
        plt.plot(self.losshist,label=["Sum","PDE","BC1","BC2","BC3"])
        plt.legend()
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.yscale("log")
        plt.title(f"Epoch = {self.epochs}")
        plt.show()


net = Net()
net.to(device)
print(net)
```
